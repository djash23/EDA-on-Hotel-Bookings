## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Analyzing the Null Values in the whole Dataframe
sns.heatmap(df.isna())
```

```python
# Removing the column with most of the null values
df.pop('company')
```

```python
# Removing remaining Null Values
df.dropna(subset=['country','children','agent','adr'], inplace=True)
```

```python
# Converting column 'children' to int
df['children']=df['children'].astype(int)
```

```python
# Reanalyzing for any other null values
sns.heatmap(df.isna())
```

```python
# Checking Sum of Null values in the columns
df.isnull().sum().sum()
```

```python
# Convert 'arrival_date_month' to categorical to ensure proper sorting
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=[
    'January', 'February', 'March', 'April', 'May', 'June', 'July',
    'August', 'September', 'October', 'November', 'December'], ordered=True)
```

```python
# Create new column which will represent total number of guests in a particular booking
df['total_guests'] = df[['adults', 'children', 'babies']].sum(axis=1)
```

```python
# Create new column which will represent total number of days stayed in a particular booking
df['total_days_stay']=df[['stays_in_week_nights','stays_in_weekend_nights']].sum(axis=1)
```

3.) Then we have removed remaining null values from different columns.

4.)Converted 'arrival_date_month' to categorical to ensure proper sorting.

5.)Created two new columns which will represent total number of guests in a particular booking and total number of days stayed in a booking.

#### Chart - 1 - Line Plot

```python
# Group by month,year and calculate total bookings year wise
monthly_bookings = df.groupby(['arrival_date_month','arrival_date_year'])['hotel'].count()
monthly_bookings = monthly_bookings.reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='arrival_date_month',y='hotel',hue='arrival_date_year',data=monthly_bookings,style='arrival_date_year',markers=True,dashes=False,palette=sns.color_palette("husl", 5))
plt.title('Monthly Booking Patterns Year Wise')
plt.xlabel('Month')
plt.ylabel('Total Bookings')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

The purpose of plotting a line chart, specifically for total monthly bookings year-wise, is to visualize the trend or pattern of booking volumes over time. By plotting total monthly bookings for each year on a line chart, you can observe how booking volumes fluctuate from month to month across different years.

##### 2. What is/are the insight(s) found from the chart?

1.) Seasonal Patterns: The chart shows if there are months when more people tend to book hotel rooms and months when fewer people book. This helps hotel managers know when they might need more staff and when they could offer discounts to attract more guests.

2.) Yearly Trends: Looking at the lines on the chart, we can see if bookings are generally going up or down each year. This tells us if more people are booking hotel rooms over time or if fewer are. It helps hotel managers understand if their business is growing or not.

3.) Impact of External Factors: Sometimes, unexpected things can affect booking numbers, like economic changes or big events in the area. The chart helps us see if these outside factors are influencing booking trends.

## **6. ML Model Implementation**

#### Random Forest Classifier Model

```python
#importing dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Load the data
path='/content/drive/MyDrive/Clean_Hotel_Booking_Analysis.csv'
clean_hotel_data = pd.read_csv(path)

# Define the target variable and features
target = 'is_canceled'
features = clean_hotel_data.drop(columns=[target])

# Identify categorical and numerical features
categorical_features = features.select_dtypes(include=['object']).columns
numerical_features = features.select_dtypes(exclude=['object']).columns

# Handle missing values for numerical features
numerical_imputer = SimpleImputer(strategy='mean')
features[numerical_features] = numerical_imputer.fit_transform(features[numerical_features])

# Handle missing values for categorical features
categorical_imputer = SimpleImputer(strategy='most_frequent')
features[categorical_features] = categorical_imputer.fit_transform(features[categorical_features])

# Encode categorical features
encoder = OneHotEncoder(sparse=False, drop='first')
encoded_categorical = encoder.fit_transform(features[categorical_features])

# Standardize numerical features
scaler = StandardScaler()
scaled_numerical = scaler.fit_transform(features[numerical_features])

# Combine processed features
features_processed = np.hstack((scaled_numerical, encoded_categorical))

# Define the target
target = clean_hotel_data[target].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_processed, target, test_size=0.2, random_state=42)

# Check the shape of the train and test sets
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# Model training and evaluation
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
print("Confusion Matrix:\n", conf_matrix)
```

```python
# The confusion matrix helps us understand the distribution of correct and incorrect predictions.
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Canceled', 'Canceled'], yticklabels=['Not Canceled', 'Canceled'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
